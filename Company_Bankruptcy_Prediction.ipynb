## Removing one column amongst columns having correlation coefficient greater than 0.9

```python
related_attributes = []
for i in range(len(cor_matrix.columns)):
    for j in range(len(cor_matrix.index) - i):
        if cor_matrix.iloc[i,j] > 0.9:
            related_attributes.append((cor_matrix.columns[i],cor_matrix.index[j]))
related_attributes
```

```python
def handle_Multico_Corr(DATA,sl=0.9):

    """[summary]
        DESCRIPTION :-
            This Methoud could be used on large or small both type of dataset

        PARAMETERS :-
            DATA = Pandas DataFrame
            sl = Columns with Corr > than 0.7 (default) will be removed
        Returns:

            Updated DataFrame after removing Multicollniearity

    """


    x_df=DATA
    corr=x_df.corr()
    columns = np.full((corr.shape[0],), True, dtype=bool)

    for i in range(corr.shape[0]):
        for j in range(i + 1, corr.shape[0]):
            if abs(corr.iloc[i, j]) >= sl:
                if columns[j]:
                    columns[j] = False

    selected_columns = x_df.columns[columns]


    return x_df[selected_columns]
```

```python
import time
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# Logistic Regression model
rf_ffs = RandomForestClassifier(criterion =  'gini', max_features = 'sqrt', n_estimators =  300)

# Forward Feature Selection
start_time = time.time()
sfs_forward = SFS(rf_ffs,
          k_features=10,
          forward=True,
          floating=False,
          scoring='accuracy',
          cv=3)

sfs_forward = sfs_forward.fit(X_train, y_train)
end_time = time.time()

selected_features_forward = list(sfs_forward.k_feature_idx_)
print("Selected features (Forward):", selected_features_forward)
print("Total time for Forward Feature Selection: {:.2f} seconds".format(end_time - start_time))
```

## Random Forest with selecting 10 features only using Backward feature Elimination

```python
import time
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

# RandomForestClassifier model
rf_bfe = RandomForestClassifier(criterion =  'gini', max_features = 'sqrt', n_estimators =  300)

# Backward Feature Elimination
start_time = time.time()
sfs_backward = SFS(rf_bfe,
                   k_features=10,
                   forward=False,
                   floating=False,
                   scoring='accuracy',
                   cv=3)

sfs_backward = sfs_backward.fit(X_train, y_train)
end_time = time.time()

selected_features_backward = list(sfs_backward.k_feature_idx_)
print("Selected features (Backward):", selected_features_backward)
print("Total time for Backward Feature Elimination: {:.2f} seconds".format(end_time - start_time))
```
